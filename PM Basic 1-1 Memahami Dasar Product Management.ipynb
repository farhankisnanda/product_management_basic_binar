Asal Usul Product Management
1. 1931: Berawal dari Memo McElroy
Neil H. McElroy bekerja di sebuah perusahaan yang namanya Procter & Gamble (P&G) yang bergerak di bidang Fast-moving consumer good (FMCG). Dia membutuhkan seseorang yang mempunyai tanggung jawab yang kompleks. Oleh sebab itu, dia menuangkan gagasannya ke dalam catatan memo dan ia ajukan ke perusahaan. Mc Elroy ngusulin posisi baru yang disebut Brand Manager, dengan tanggung jawab:
- Cukup mengelola satu brand aja
- Bisa dekat dengan customer
- Bertanggung jawab secara menyuluruh sama proses dari brand tersebut, mulai dari pengembangan produk, analisa penjualan dengan cermat, sampai pemaparan hasil analisanya ke atasan.
Kenapa McElroy ngusulin ini? Karena tiap brand yang dimiliki perusahaan masih dikelola secara terpusat, alhasil kinerja perusahaan masih dikelola secara terpusat. Pada waktu itu, brand juga masih mengesampingkan suara customer. Padahal, menurut McElroy, ini penting banget biar bisa ngembangin brand lebih baik lagi. Dan usulan ini disambut baik oleh perusahaan dan malah McElroy jadi pelopor lahirnya product Management.
2. Dikembangkan oleh HP
Dimasa tua McElroy, dia ngajar di Stanford University, terus ketemu sama pengusaha muda Bill Hewlett dan David Packard (co-founder HP). Mereka belajar dari McElroy dan memutuskan pada tahun 1943 HP membuat tim terpisah yang khusus buat fokus ke produk masing-masing. Tiap tim bertugas buat mengembangkan, memproduksi, dan melakukan marketing untuk produknya sendiri. Nah, tiap tim ini dipimpin sama yang namanya Product Leader.
3. 1848-1975: Sistem yang Serupa oleh Toyota
Selain McElroy yang jadi pencetus product management, ternyata muncul juga di Jepang. Setelah perang dunia ke-2 berakhir, Jepang kekurangan dana ada kendala cashflow. Gara-gara ini, semua industri di Jepang dituntut baut bisa just-in-time manufacturing alias harus serba hemat dan efisien. Di masa itu, Perusahaan Toyota ngga sekedar mau hemat aja. Mereka mau lebih dari itu. Merek mikirin gimana biar manufacturing juga bikin perusahaan bisa scale-up. Akhirnya terciptalah visi baru Toyota yang disebut The Toyota Way. Apa aja visinya?
- Kaizen: Fokus buat improvement terus-menerus sambil tetep inovasi.
- Genchi Genbutsu: Berpegang pada fakta dan data buat bikin keputusan.
Visi ini diterapkan oleh Toyota selama hampir 30 tahun, dari 1948 hingga 1975 dengan sistem produksi yang dikepalai Product Manager.
4. Munculnya Agile
Sebelum adanya Agile, masih banyak perusahaan yang ngembangin produk dengan metode waterfall. Oleh sebab itu, munculla metode Agile supaya perusahaan tetap bisa berkreasi dan ngeluarin produk yang relevan sama marketnya. Metode ini awalnya tercetus dari konspen Agile Manifesto, buat ngembangin software. Apa aja isinya?
- Individu dan interaksi lebih penting dari pada proses dan alat.
- Software yang berfungsi lebih penting dari pada dokumentasi yang komprehensif
- Kolaborasi pelanggan lebih penting dari pada negoisasi kontrak
- Menanggapi perubahan lebih penting daripada rencana
Lahirnya Agile Manifesto ini berperan penting di dunia product development modern. Agile membantu melancarkan hubungan antara tim product dan engineer dan memperjelas visi product management itu sendiri. Selain itu, dengan metode Agile, jadi nggak ada sekat pemisah lagi antara tahap research, spesifikasi, dan development.
5. Gimana perkembangan Product Management sekarang?
Pada dasarnya, perkembangan product management ini membentuk 4 landasan utama dalam praktik product management modern:
- Berinteraksi dengan konsumen dan mengungkap masalah mereka.
- Mengembangkan produk yang memecahkan masalah tersebut
- Membuat strategi penjualan dan jalur distribusinya untuk menjual produk
- Menganalisa metrik yang tepat, meneruskan iterasi, dan mendorong terjadinya profit.
Visi The Toyota Way berkembang menjadi metode baru yang disebut Lean. Metode Lean ini bertujuan buat mastiin bahwa produk dan perusahaan selalu berinovasi dan mendengarkan kebutuhan customer. Caranya dengan continous improvement, user testing, dan monitoring metriks-metriks pengembangan produk secara berkala.
Glosarry:
- Agile Manifesto adalah nilai-nilai software development hasil gagasan 17 software development yang menjadi dasar metode Agile.
- Learn adalah metode peni9ngkatan proses kerja agar lebih efektif, efisien, dan berkualitas dengan menghilangkan aktivitas yang tidak punya nilai tambah.
## Posisi & Tanggung Jawab Product Manager

### Tanggung Jawab Product Manager

- PM disebut sebagai middleman karena dia menjadi penghubung antara tim internal, customer, dan stakeholder. 
- Adapun peran PM ialah: Mencari tahu masalah yang ada agar tahu apa yang perlu dibenahi dan dikembangkan dari produk, mengelola produk secara kseluruhan. menentukan spesifikasi produk secara rinci.
- Adapun perbedaan Prodcut Manager (PM) dengan Product Owner (PO) ialah PM berfokus ke strategi pengmbangan produk, PO berfokus ke pengembangan secara teknis.
Secara garis besar, tanggung jawab seorang Product Manager (PM) adalah sebagai berikut:
1. Own th product, bertanggung jawab penuh terhadap produk (merancang, membaut strategi, dan visi product)
2. Mengumpulkan problem-problem dan mencari pokok permasalahannya.
3. Membuat strategi, roadmap, dan spesifikasi produk dengan rinci.
4. Berkolaborasi sama tim lain buat mengembangkan produk.
5. Memantau performa produk dan menentukan area yang perlu dibenahi dari produk (feedback user, jadwal market expansiom)
Perbedaan antara Product Manager, Product Owner, dan Scrum Master
Product Owner berfokus pada technical yang lebih berurusan dengan Tim Tech dan Tim UI-UX, sedangkan Product manager berfokus pada strateginya dan lebih berurusan dengan rancangan strategi bisnis, budget planning, tim marekting, dan pemahaman market. Namun ngga jarang dua role ini disatukan karena dua"nya memang berfokus dalam memantau jalannya pengembangan produk. Awal mula, pembangunan startup, tentu CEO akan menjadi PM sekaligus. Seorang PM tidak harus jago diberbagai bidang, yang terpenting paham proses bisnis, paham flow engineering, dan paham proses desain. Yang mana fungsi utamanya memahami berbagai bidang lain adalah bisa berkomunikasi dengan lancar ke tim-tim terkait. PM harus memahami perspektif seluruh tim terkait. PM bukan berarti leader karena setiap tim punya leadernya masing". Oleh sebab itu, perlu menjalin hubungan baik dengan member tim. Adapun Scrum master, hanya memantau metode scrum dan agile yang berjalan.

## Tugas, Peran, & Skill Product Manager

### Tiga Tugas Utama Product Manager

1. Connecting and Communicating, caranya degan melakukan meeting secara rutin dengan anggota tim dan stakeholder. Harus bisa menjadi a good communicator.
2. Learning and Analytical Thinking, Fokus pekerjaan Product Manager adalah product development. Dalam mengembangkan product PM haru mempelajari dan menganalisis produk yang ia handle. Apa saja yang harus dianalisi, banya banget. Mulai dari mempelajari penggunaan produknya kayak gimana, siapa target marketnya, analisi kompetitor, dan menentukan fitur produknya mau kayak gimana. Selanjutnya PM juga harus measuring produk. Apa saja itu? PM harus menganalisis Key perfomance Index (KPI) produk, menganalisis campaign yang udah berjalan, dan review fitur-fitur dari produk-bagaimana tanggapan customer. 
3. Pengarah, Pembuat Keputusan, dan Dokumentasi, PM harus bisa mengarahkan strategi ke arah yang selaras dengan kebutuhan customer, kepentingan bisnis, dan internal, dan juga harus memutuskan keputusan akhir jika diskusi tak kunjung berakhir. Nggak lupa juga, Product Manager (PM) harus mendokumentasikan:
- Selalu membuat summary dari semua meeting.
- Menuliskan permasalahan dan hipotesis kepada stakeholder terkait.
- Membuat report berkala kepada stakeholders
- Review & grooming Backlog secara rutin
Glossary:
- Key Perfomance Index (KPI): Metrics finansial maupun non-finansial yang digunakan untuk membantu suatu organisasi menentukan dan mengukur kemajuan terhadap target capaian perusahaan.

### Tiga Skill Utama Product Manager
1. Communication Skill, PM harus berpikiran terbuka dan lebih banyak mendengar. Karena terkadang user tidak bisa menceritakan secara jelas keinginannya. 
2. Management Skill, selain bertugas untuk memastikan bahwa fungsi bisnis dan produk berjalan dengan selaras, PM juga bertugas untuk memastikan bahwa waktu pengerjaan dari strategi produk dapat berjalan dengan lancar.
3. Technical Skill, yang dibagi menjadi 3:
- Bisnis, seorang PM harus menguasai bisnis dan market. Ia harus dapat menegrti siklus pengembangan produk bahkan sampai ke level bisnis.
- User Experience, seorang PM harus mengerti tentang user experience untuk memastikan bahwa solusi dan produk yang dikembangkan memiliki user experience yang baik.
- Tech, PM tidak wajib untuk mengetahui bagaimana cara memprogram, namun setidaknya PM harus mengetahui arsitektur sistem dan bagaimana sistem berkomunikasi.
