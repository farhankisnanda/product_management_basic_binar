## Memahami Business Model Perusahaan

### Apa Itu Bisnis Model?

Business model merupakan gambaran bagiamana sebuah perushaan dapat berjalan dan mengambil keuntungan. Dari sini pulalah kita dapat melihat bagaimana perusahaan beroperasi, menjalankan bisnisnya, menghasilkan produknya dan mendapatkan keuntungan. Bisnis artinya kegiatan yang berkaitan dengan transaksi jual-beli barang atau jasa serta bertujuan untuk memperoleh keuntungan. Dari business model, kita bisa melihat denah dan bagaiamna perusahaan bekerja. 

### Menggunakan Business Model Canvas

Business model penjulan apel misalnya, menggambarkan:
- Siapa supplier apelnya?
- Bagaimana cara mengambil apel untuk dijual di toko?
- Bagaimana mencari pembelinya?
- Bagaimana membuat pembeli setia dengan toko?
- Darimana keuntungan didapat?
- Apakah dapat keuntungan dengan jual apel saja?
- Apakah apelnya bisa diolah jadi jus untuk dapat keuntungan lebih?
- Apakah akan jai distributor untuk toko-toko yang kecil?
Business Model membantu kita untuk mengembangkan produk sesuai arahan bisnis perusahaan. Pembuatnya adalah Alexander Osterwalder. 

### Lebih Dalam Soal Business Model Canvas

1. Customer Segment, adalah keompok target yang akan atau sedang kita bidik untuk menjadi customer perusahaan. Customer bisa dikelompokkan berdasarkan usia, minat, kebiasaan belanja, dan lainnya. Kita harus tau berapa banyak customer segment dalam produk kita. Yang mana nantinya customer segmnet ini bisa diubah jadi user persona bisnis. Kita harus dapat menjawab pertanyaan:
- Produkmu dibuat untuk memecahkan masalah siapa?
- Siapa customer yang mungkin menyukai produkmu?
- Apakah mereka pake produk yang sama dari brand lainnya?
- Apakah produkmu disukai wamita atau pria, atau bahkan keduanya?
- Berapa usia customer yang tertarik dengan produkmu?
2. Value proposition, kira perlu mengidentifikasi nilai atau value utama yang ingin kita tawarkan ke customer. Kita dapat menulis kelebihan produk kita dari produk-produk serupa yang bisa menjadi perhatian dan menjawab kebutuhan customer. Berikut ada beberapa contoh kategoriu value yang bisa jadi merupakan keunggulan perusahaan atau produk kamu:
- Newness, produk/jasa yang belum pernah ditemukan sebelumya.
- Perfomance, produk/jasa yang ditawarklan meningkatkan kinerja customer jadi lebih efisien/efektif.
- Cuatomization, produk/.jasa yang ditawarkan berbeda atau ada pilihan untuk setiap segmen.
- Getting the job done, membantu untuk menyelesaikan sesuatu.
- Design, menawarkan nilai artisitik lebih dari sedar fungsional.
- Brand, brand high class memberikan status kepada pembelinya.
- Price, menawarkan harga yang bersaing atau sesuai dengan karakteristik segmen pasar.
- Cost reduction, membantu customer mengefisienkan biaya pemakaian.
- Risk reduction, menawarkan produk/jasa yang meninimalkan risiko ditanggung customer kayak garansi.
- Accessibility, memudahkan akses customer terhadap produk/jsa yang ditawarkan.
- Convenience, menawarkan produk/jasa yang memudahkan customer.
3. Channels, kita juga harus mikin gimana cara yang tepat buat menjangkau customer. Mulai dari pendistribusian sampai produkmu ada ditangan customer. Mulai dari pendistribusian sampai produkmu ada ditangan customer.
4. Customer relationship. Bedanya dengan channel, customer relationship berkaitan dengan hubungan, sedangkan channels berkaitan dengan cara menjangkau customer. Secara umum ada 7 jenis hubungan bisnis sama customernya. Kita dapat pake lebih dari 1 hubungan yaitu:
- Transactional, hubungan sebatas transaksi pada saat itu aja, misal beli kopi pas berhenti di rest area.
- Long-term, hubungan yang deep dan long-term, misalnya antara dokter gigi dan client-nya.
- Personal assistance, customer bisa berinteraksi langsung dengan representatif perusahaan, misal staf gerai skincare yang bantuin customer pas belanja.
- Self service, customer bisa mandiri tanpa kita arahkan, misalnya di supermarket.
- Automated service, layanan bisnis kita otomatis, misalnya penonton netflix bisa daftar sampai dapat rekomendasi dilm nggak kita pada, cukup pake aplikasi aja.
- Community, menciptakan komnitas buat customer, misalnya Honda bikin komunitas yang rutin ngadain liga sepakbola dan riding bareng.
- Co-Creation, hubungan yang ngajak customer buat ciptain sesuatu di bisnismu, contohnya Youtube.
5. Revenue stream, cara untuk menghaslkan keuntungan dari value proposition kita. Dapat berupa feature listing, advertising, sales commision, subscription, dan affiliate.
6. Key resources, sumber daya buat ngejalanin bisnis kita. Misalnya mulai dari ruang kantor, peralatan kantor, transportasi, dampe para karyawan.
7. Key activities, kita harus mikirin gimana caranya agar produkmu bisa mencapai nilai/value yang diinginkan customer. Mulai dari mikirin strategis bisnis, distribusi produk, ningkatin keahlian karyawan, sampe bahan baku produkmu.
8. Key patners, kita juga perlu mikirin pihak-pihak yang akan diajak kerja sama, agar bisnis kita bisa semakin berkembnag dan mencapai value yang diinginkan customer.
9. Cost structure, kita juga harus memperkirakan rincian biaya yangharus dikeluarkan. Dari mengidentifikasi key resources, key activities, dan key partners, kamu juga jadi bisa menjabarkan biaya-biaya yang dibutuhkan buat ngehasilin produk atau jasa kita dengan value proposition yang sudah ditentukan tadi.
Glosarium:
- Feature listing, cara buat meng-highlight produk di tampilan depan website.
- Advertising, periklanan atau jenis pemsaran berbayar,
- Sales commission, cara perusahaan buat mendorong karyawan biar berusaha sebaik mungkin buat jual produk; kompensasi fisik atau nonfisik.

## Memahami Struktur Organisasi Perusahaan 

### Beberapa Macam Struktur Organisasi

<img src="https://i.ibb.co/5xWqBqq/Whats-App-Image-2022-07-25-at-7-39-44-AM.jpg" width="500">

1. Vertical structure, kayak hirarki bersusun yang makin tinggi letaknya, makin tinggi juga tanggung jawabnya. Struktur organisasi tipe ini cocok buat diterapin di perusahaan yang udah stabil, yang kontrol manajerialnya kuat dari tingkatan atas ke tingkatan bawah karena tujuannya biar menimalisir kesalahan. Tapi, kalausebauh perusahaan pake struktur ini, efeknya ketika ada perubahaan prosesnya bakal lama, soalnya hierarki yang tinggi. Buat ngambil sebauh keputusan juga harus ngelewatin jalur yang panjang dan kurang fleksbel.

<img src="https://i.ibb.co/rGpX6R6/Whats-App-Image-2022-07-25-at-7-39-43-AM.jpg" width="500">

2. Horizontal structure, biasanya dipake oleh perushaan yang baru berkembang atau masih negrintis perusahaannya, jadi tim sama anggotanya juga masih sedikit. 
Kalo berdasarkan kultur dan gaya komunikasinya secara aplikatif di perusahaan umumnya sturuktur organisasi itu dibagi menjadi organic structure dan mechanistic structure.

<img src="https://i.ibb.co/QKWwKzL/Whats-App-Image-2022-07-25-at-7-39-43-AM-1.jpg" width="500">

3. Organic structure, dasarnya adalah horizontal structure. Struktur ini fleksibel banget, setiap orang bisamenyesuaikan kerjaanya tergantung keperluan di saat-saat tertentu. Jaringan informasi juga jadi lebih luas dan PM jadi lebih fleksbel buat ngambil keputusan , ngeapprove produk baru, atau adaptasi teknologi baru.

<img src="https://i.ibb.co/3pwTvJ7/Whats-App-Image-2022-07-25-at-7-39-43-AM-2.jpg" width="500">

4. Mechanisitic structure, dasarnya adalah vertical structure. Struktur ini birokratis dan lebih terpusat, pembagian tugasnya juga udah dikasih dengan jelas, jadi anggota timnya cuma perlu ngikutin apa yang udah diperintahin sama atasan. Pengambilan keputusannya juga cuma dilakukan sama kepala atau ketua timnya aja, jadi prosesnya relatif lebih terstruktur karena udah ada aturan yang jelas.


## Memahami Metrics Perusahaan

### Goal-Setting dengan OKR

<img src="https://i.ibb.co/N1pH111/Whats-App-Image-2022-07-25-at-10-24-17-AM-1.jpg" width="500">
<img src="https://i.ibb.co/MPvtJTX/Whats-App-Image-2022-07-25-at-10-24-18-AM.jpg" width="500">

Secara definisi, Objective dan Key Results adalah pendekatan yang dipake buat menyelaraskan tujuan dari semua bagian perusahaan. Mulai dari cakupan yang paling luas, sampai yang paling sempit. Setidaknya ada 4 level dari OKR ini, yaitu:
1. Global, OKR dari perusahaan secara keseluruhan.
2. Units, OKR dari departemen.
3. Teams, OKR suatu tim.
4. Person, OKR individu.
Berikut penjelasan bagaimana OKR bisa ngebantu perusahaan:
1. OKR bikin seluruh anggota perusahaan jadi punya tujuan yang selaras, bahwa seluruh bagian perushaan bekerja ke arah yang sama.
2. OKR membantu perusahaan buat melakukan prioritas task dengan cara memfokuskan kegiatan yang memiliki dampak paling besar terhadap bisnis.
3. OKR membantu seluruh anggota prusahaan mengerti dengan jelas tujuan dan strategi bisnis perusahaan.
Selain itu, OKR punya tujuan juga untuk ngejawab pertanyaan berikut:
1. Kemana tujuan kita (objective)? Kita bisa identifikasi 3-5 goals dari perusahaan, tim atau indivdu. Goals dari OKR ini harus ambisius, puynya target waktu tertentu, namun realstis, artinya dapat dicapai sampai level tim/individu.
2. Apa tolak ukur tujuan kita (key result)? Selanjutnyakita perlu negjabarin cara ngukur goals yang mau dicapai. Kita bisa nentuin 4-5 key result dari masing-masing goals. Yang terpenting, elemen key result yang kita susun bisa siukur, dicapai, punya objective yang sulit, tapi tetap harus realistis.
Glosarium:
- Metrics: alat ukur kinerja karyawan.
- North Star Metrics: alat ukur yang jadi pertunjuk arah yang tetap. Metrics utama yang menajdi fokus dan yang paling berarti untuk perusahaan, karena berhubungan denkat dengan goal perusahaan.
- Key result: hasil utama atau hasil yang bisa ngukur tujuan perusahaan.

### Memonitor Performa dengan KPI

<img src="https://i.ibb.co/0MSjzYF/Whats-App-Image-2022-07-25-at-10-24-17-AM.jpg" width="500">

Key Perfomance Indicator, sebuah aktivitas yang mendukung buat mencapai goals atau target yang di dalemnya ada langkah-langkah tertentu biar nantinya bsia diukur. Dengan kata lain mengukur sejauh mana sebuah perusahaan mencapai objective bisnisnya. Dan juga sebagai pengukur kinerja tim atau individu buat mencapai target perusahaan yang udah ditentuin dan dijadiin sebagai tanda apakah perusahaan masih sehat dan mampu untuk mencapai tujuan tersebut atau nggak. Singkatnya peran kita sebagai PM adalah ngejalanin key yang udah dibuat dan nanti bakal dinilai proses dan hasilnya pake indikator pengukuran perusahaan berupa KPI. KPI punya beberapa karakteristik diantaranya:
- Menggunakan ukuran yang paling sering dipake.
- Menggunakan ukuran yang diketahui oleh manajemen. Supaya kalau ada sesuatu diluar perencanaan, bisa dimonitor dan langsung ambil tindakan selanjutnya.
- Bisa dipahami oleh semua orang di tim.
- Sesuai tanggung jawab masing-masing. Meskipun perusahaan punya KPI khusus, tapi KPI tim tech dan tim bisnis bisa beda karena tugasnya pun beda. 
- Punya dampak yang signifikan.
- Punya dampak yang positif. 
Lalu, apa hubungannya goals, OKR, dan KPI? Kalau dianalogiin sebagai perjalanan, goals, OKR, dan KPI ini punya peran yang saling berkaitan. Misalnya, kamu mau pergi perjalanan ke Monas, berarti Monas ini adalah goals-nya. Nah buat ngebantu kamu nemuin lokasi yang sesuai sama Monas, kamu bisa pake GPS atau Maps, mereka ini adalah OKR yang bakal nunjukin jalan mana yang mau kamu pilih, mau jalan utama atau jalan alternatif. Terus selama di perjalanan kamu sering banget berenti buat istirahat sekaligus isi bahan bakar. Nah KPI inilah yang bakal nilai gimana proses perjalanan kamu. Yang harusnya tujuan Monas bisa ditempuh 2 jam dan bahan bakarnya cukup, tapi karena kamu sering berenti dan suka kelewatan pas mau belok, jadi dalam waktu 3 jam kamu baru nyampe Monas. Nantinya KPI bakal nilai kalau waktu perjalanan yang kamu pake buat ke Monas ini masih kurang efektif. Goals adalah target, OKR adalah GPS-nya supaya bisan sampai ke target. Sedangkan KPI adalah pengukuran setelah proses selesai.

## Memahami Change Management

### Apa Sih Change Management?

Change management merupakan pendekatan terstruktur untuk nerapin perubahan, mulai dari menyusun strategi hingga monitoring. Pendekatan perubahannya bisa berupa inovsi ataupun menyesuaikan perubahan yang udah ada supaya operasional perushaaan lebih efisien. Berikut beberapa alasan penerapannya:
1. Mempertahankan keberlangsungan hidup perusahaan, baik dari eksternal seperti blackberry yang tidak mau berubah ke touchscreen dan juga internal seperti mempertahankan hirarki perusahaaan yang membuat perusahaan sulit berkembang.
2. Menyesuaikan dengan perubahan yang telah terjadi di perusahaan. Seperti merubaha dari gaya waterfall ke agile-scrum.
3. Memperbaiki pola efektivitas perusahaan, seperti mengurangi waktu meeting yang ternyata kurang efektif.

### Bagaimana Cara Mengelola Perusahaan?

1. Identifikasi formulasi perubahan, diantaranya:
- Mastiin semua pihak di perusahaan paham kenapa kita nerapin perubahan. Biar semua sepakat dan satu pemahaman.
- Bikin visi perusahaan dan mastiin stakeholders paham hasil peruabahan yang diharapkan. Kalau bisa jangan ada ruang untuk ambigu.
- Mastiin perusahaan paham siapa aja yang jadi agen perubahan.
2. Perencanaan perubahan, berikut detailnya:
- Kalau produk yang kita pegang berkaitan langsung dengan user, diskusikan sama tim marketing buat rencanain perubahan bareng-bareng.
- Buat timeline yang terstruktur, berikan detailnya.
- Project besar dipecah jadi proses-proses yang lebih kecil biar lebih jelas dan kebayang.
- Sampein rencana perubahan ke stakeholders terkait seperti CEO, tim UX, dan tim engineer.
3. Implementasi perubahan, harus dimonitoring dan apresiasi progres-progres kecil anggota tim, ditraktir mungkin.
4. Tahap evaluasi dan feedback, diperlukan data yang relevan untuk jadi landasan penilaian atas perubahan.
