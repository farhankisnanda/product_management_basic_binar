## Empati


### Memahami Empati

Empati adalah menempatkan diri menurut sudut pandang orang lain. Adapun simpati adalah respons emosional yang kita rasakan, simpati tidak berupaya menempatkan posisi kita ditempat orang lain. Cara empati dapat berupa; diam sejenak, amati, dengarkan, pahami kondisi, pertimbangkan keputusan. Roman Krzanic salah satu silfus menyimpulkan bahwa Empati adalah menempatkan diri menurut sundung pandang orang lain, memahami apa yang orang lain butuhkan dan rasakan terlepas dari sudut pandang kita sendiri.

### Product Manager: Mulai dari Empati ke Pengguna sampai ke Diri

1. Empati dengan Pengguna, seorang PM harus tau dulu beberapa hal tentang para pengguna produk mereka, misalnya; Apa sih kesulitan yang mereka hadapi? Atau, butuh waktu berapa lama buat mereka menggunakan produk tersebut?
2. Empati dengan Customer. Customer adalah mereka yang bayar untuk menggunakan produk. Sedangkan pengguna atau consumer adalah mereka yang menggunakan produknya. Contohnya sebuah produk mainan anak-anak tentu nggak dipake sama orangtua. Walaupun demikian, orang tua tetap harus bayar produk mainan tersebut supaya anaknya bisa bermain. Orangtua adalah customer, dan anak-anak adalah consumer.Berempati kepada customer berarti, memahami bahwa meskipun tidak menggunakan produknya seorang Product Manager tidak boleh mengurangi kebutuhan yang diharapkan customer. Misalnya terkait harga produk, batasan yang diperlukan, ataupun syarat penggunaan.
3. Empati dengan Engineering dan Tim Desain, cara nunjukin empatinya adalah dengan menjalin komunikasi yang baik dengan tim dan berdiskusi buat ngedengerin saran buat ningkatin sistem ataupun ngebahas tentang penggunaan teknologi saat ini.
4. Empati dengan Stakeholders, caranya bisa dengan dulu gimana sih masalah yang dihadapi oleh stakeholders, setelah itu PM bisa sampein juga permasalahan yang dihadepin.
5. Empati dengan Perusahaan, seorang PM perlu mastiin visi dan misi perusahaan sejalan dengan proses pengembangan produk. Dengan kata lain, produk yang baik akan tercemin apabila Product Manager berhasil menjembatani value perushaan dengan kebutuhan customer. Apakah pendekatan ini sesuai? Apakah ide pribadi sejalan dengan budaya perusahaan?
6. Empati dengan Diri Sendiri, seorang PM harus memahami bahwa tujuan, ambisi, kegagalan, dan pembelajaran yang selama ini dihadapi bisa jadi sumber kesempatan buat memperbaiki di masa depan.

### Summary 4 Langkah Empathy

1. Jadi pendengar yang baik, Usahakan ngobrol dengan customer minimal sekali dalam seminggu. Memang kita bisa meminta data kepada tim sales atau liat data anlytics dari product metrics. Hanya saja perlu diperhatikan, data bisa membantu kita, tapi raw numbers nggak selalu bisa menunjukkan gambar menyuluruh.
2. Jangan menghakimi!, menilai suatu kondisi hanya dari sudut pandang pribadi dapat mengurangi rasa empati. Do not let our confidence in our own direction allow you to put up blinders to other solutions that might work.
3. Bersikap membuka diri, tapi tetep tau batasan
4. Bertanya untuk memahami kebutuhan, seperti mengapa bapak memerlukan barang ini? Mungkin saja, kita dapat memberikan solusi yang lebih wow untuk dia. 
5. Menggunakan produk kita sendiri,seortang Product Manager tidak akan bisa berempati pada permasalahan yang tidak ia mengerti. 

## Problem Solving: Six Thinking Hats

### Apa Itu 6 Thinking Hats?

Kemampuan Problem Solving adalah gimana cara kamu memetakan masalah dan memilih solusi yang tepat. 
### Konsep 6 Thinking Hats?

Penciptanya adalah Edward de Bono. Beliaulah yang menulis Six Thinking Hats pada tahun 1985. Six Thinking Hats ini bisa bantu nyelesain masalah baik itu masalah pribadi maupun masalah tim ketika ngambil sebuah keputusan yang didasarkan pada berbagai sudut pandang. Singkatnya teknik ini bermanfaat untuk beberapa hal yaitu:
- Ngasih inspirasi alternatif penyelesaian.
- Bikin pertemuan tim jadi lebih produktif.
- Memperbaiki keputusan tim.
Mari kita bahasa mengenai Six Thinking Hats-nya:
1. Topi Biru - Si Fasilitator, topi biru memfasilitasi dan menjaga agar proses penyelesaian masalah terorganisir dengan baik untuk mencapai solusi. Dia juga yang akan membuat keputusan dan rencana apa yang harus diambil di akhir.
2. Topi Hijau - Si Kreatif, topi hijau bisa ngehasilin ide-ide kreatif, hal-hal yang out of the box buat bahan diskusi.'
3. Topi Putih - Si Pencari Fakta, topi putih berfokus pada data dan informasi yang bersifat objektif, kita akan lebih banyak menganalisa tren-tren lama dan mengeksplorasi data-data historikal.
4. Topi Kuning - Si Positif, topi kuning berpikir positif dan melihat sisi baik dari sebuah keputusan atau alternatif solusi.
5. Topi Merah - Si Emosional, Topi merah digambarin sebagai topi yang ngeliat segala sesuatu dengan perasaan dan intuisi. 
6. Topi Hitam - Si Pesimis, Tpoi ini sering diasosiasikan dengan nunjukin segala sisi buruknya pas mau ngambil sebuah keputusan. Dengan begitu kita akan berhati-hati akannya resiko. Topi ini berfokus ke berpikir kritis dan menganalisis risiko dari tiap keputusan, sehingga seringkali terlihat pesimis dan skeptis

### Problem Solving Kelompok dengan 6 Thinking Hats

Alurnya bisa seperti berikut:
- Nentuin siapa sebagai topi biru
- Semua orang memakai topi hijau, menyampaikan ide
- Semua orang memakai topi putih, menyampaikan fakta
- Semua orang memakai topi merah, menyampaikan perasaan pribadi
- Semua orang memakai topi hitam, menyampaikan risiko
- Semua orang memakai topi kuning, menyampaikan hal positif
- Kesimpulan dan pengambilan keputusan oleh topi biru.

## Influencing Skills

### Product Manager Influencing Skill Starter Pack

Langkah-langkah PM mengasah kemampuan influencing:
1. Visi Produk yang Jelas, gampang dibayangin dan fokus sama customer.
2. Buat List Siapa Aja Orang Penting yang Harus di Influence, PM harus tahu apa yang mereka butuhin. Minimalnya, kebutuhan fungsional dan kebutuhan emosional.
3. Bangun Relasi dan Kepercayaan, rumusnya kepercayaan = kredibilitas + reliabilitas + intimasi.
- Kredibilitas, artinya seberapa meyakinkan kamu dalam memperlihatkan bahwa kamu menguasai topik atau bidang tertentu. Adapun cara membangun kredibilitas dapat melalui; Menggunakan produk sesering mungkin, lakukan competitor analysis, gunakan produk kompetitor, rutin ngobrol sama customer, liat data market biar tau titik kuat dari produkmu dan apa yang harus dikembangin.
- Reliabilitas, artinya kemampuanmu buat bikimn orang lain bergantung padamu. Caranya, dapat melalui memberikan performa yang konsisten, dan cari data pendukung biar kamu yakin bahwa informasi yang didapatkan itu bener dan no hoax. Cara membangun realibilitas dapat melalui; Jangan over promise atau kebanyakan janji, disiplin dan teratur, pastiin kesepakatan yang kamu buat punya konteks yang jelas, tinjau action items(hal-hal yang perlu dilakukan)
- Intimasi atau keakraban, artinya untuk membangun perasaan bahwa seluruh tim berjuang bersama, perlu komunikasi rutin dengan tim dan stakeholder, serta melibatkan mereka dalam proses discovery produk. Sampaikan dengan jujur ketika ada isu agar bisa dicari solusi.

## Critical Thinking: Memvalidasi Request

PM harus bisa memilah request yang masuk dan menilai beberapa aspek seperti:
1. Apakah sesuai dengan  kebutuhan user?
2. Besar kecilnya effort dalam mewujudkan hal tersebut?
3. Impact atau dampak dari request tersebut?

### 4 Cara Validasi

1. Dengarkan Keluhan dan Feedbcak dengan Saksama, seorang PM harus mendengarkan feedback dan permintaan mereka dengan saksama, berempati selama identifikasi masalah, menggunakan kemampuan active listening. Active listening dilakukan untuk membangun koneksi dengan customer atau stakeholdder dengan tidak menyela, menunjukkan body language yang baik, serta mengulang dan mengkonfirmasi request untuk memastikan. Walalupun belum tentu kamu terima, berikan kalimat feedback dengan alasan yang masuk akal dan dapat dimengerti agar customer maupun stakeholder merasa didengar.
2. Pahami Produk dan Analytics dengan Baik, mau itu resource internal, kompetitor, maupun faktor-faktor lain yang dapat mempengaruhi keputusanmu saat memilih dan mempertimbangkan.
