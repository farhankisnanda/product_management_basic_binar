- Judment & decision making, kemampuan menimbang-nimbang suatu keputusan secara subjektif dan humanisi.
- Digital Literacy, kemampuan menggunakan berbagai platform digital alias nggak gaptek.

### 2.0 Sejarah Aplikasi

Game snack adalah aplikasi pertama yang dibuat yang ada pada fitur hape Nokia. Awal mula aplikasi internet dikarenakan Steve Jobs yang menyediakan AppStore sebanyak 522 aplikasi pada tahun 2008.

### 2.1 Mobile App vs Mobile Website

Mobile web itu adalah website yang dapat diakses lewat mobile device. Adapun mobile app adalah program berbentuk aplikasi yang bisa diakses lewat mobile device.

|                  | Mobile Web | Mobile App | 
| :--------------- | :----------- | :---------- |
| Akses            | Bisa diakses dengan mengetik URL di browser | Perlu di dowload melalui Google Play Store atau Apple Store |
| Penyimpana data  | Tersimpan di server online, sehingga tidak menghabiskan <br> storage di mobile device | Karena perlu di download maka akan memakan memori |
| Updates          | Terjadi secara otomatis, user nggak perlu mengupgrade | User akan diberi notifikasi dan diminta untuk upgrade |
| Kompatibilitas   | Bisa diakses di semua mobile device dan operatisng sistem | Umumnya punya syarat device atau operating sistem tertentu |
| Sambungan fitur device | Nggak tersambung langsung dengan fitur device | Fitur aplikasi tersambung dengan device |
| Fleksibilitas | Kurang fleksibel digunakan sehari-hari | Lebih fleksibel |
| Proses development | Development website lebih hemat, cepat, dan mudah | Develpoment aplikasi perlu waktu dan biaya yang lebih besar |

Dan juga salah satu benefit yang terpenting dari menggunakan mobile app, yaitu memberikan kemudahan dan security.

## 3. Memahami Konsep Produk

### 3.1 Product vs Project

Facebook verasal dari project iseng bernama Facemash. Di project tersebut Mark Zuckerberg hanya punya satu tujuan, yaitu bikin platform kecil-kecilan yang nampilin foto orang secara random. Menurut Project Management Institute (PMI), project adalah sesuatu yang bersifat sementara, karena memiliki awal dan akhir yang telah ditentukan. Selain itu project juga cuma memiliki satu tujuan urtama sehingga kalau tujuannya udah tercapai, ya berarti projectnya beres atau bahasa kerennya one-off delivery.  

Adapun product adalah barang, jasa, sistem, atau platform yang dibuat dengan dua tujuan utama yaitu menyelesaikan masalah banyak orang dan menguntungkan perusahaan dalam jangka panjang. Jadi product ini lebih besar dan lebih komplek dibandingkan project. Maka dari itu, dalam pengembangannya, product bisa dipecah menjadi rangkaian project. Dan bedanya product bersidat continous improvement. 

| Product | Project |
| :------ | :------ |
| Tujuan lebih kompleks | Tujuan bersifat tunggal |
| Pengerjaannya permanen kecuali dinonaktifkan | Ada tanggal dimulai dan tanggal berakhir | 
| Timnya jangka panjang | Timnya jangka pendek |
| Continous improvements | One-off delivery |
| Berkembang sesuai kebutuhan bisnis dan customer | Berkembang sesuai persyaratan project |

Adapun keuntungan kebanyak produk teknolohi mendapatkan revenue (pendapatan) dari data user. 

### 3.2 Macam-Macam Revenue Stream

Revenue adalah pendapatan, adapun stream adalah aliran pendapatan. Revenue stream dapat diartikan sebagai sumber perusahaan untuk mendapatkan pendapatan yang berkelanjutan.  

Contoh revenue stream: 
1. Direct Sales, alias penjualan barang secara langsung dan sekali putus (cukup bayar sekali di awal dan selanjutnya tidak perlu lagi).
2. Fremium, terdapat beberapa fitur yang dapat digunakan dengan berbayar.
3. Advertising
4. Sales Commision, perusahaan bekerja sama dengan pihak tertentu buat berbagi hasil pendapatan. Seperti bagi hasil di Go-Food 80% untuk restoran dan 20% + 1000 Go-Food.
5. Subscription(Langganan), berbayar perbulan agar dapat memiliki akses, contohnya Netflix.

### 3.3 Frontend dan Backend Side dalam Aplikasi

Frontend adalah bagian yang menghubungkan sebuah aplikasi dengan user. Caranya ialah dengan menampilkan interface atau grafis yang dibutuhkan user saat menggunakan aplikasi. Frontend memastikan agar isi aplikasi seperti; tombol, image, teks, dan sebagainya bisa interaktif. Adapun Backend adalah bagian tempat menyimpan serta memproses data. Supaya user bisa input dan request data.  

## 4. Memahami SDLC

### 4.1 Pendekatan Waterfall dan Agile

SDLC (Software development Life Cycle) ialah suatu pendekatan baut ngatur proses dan pembagian tugas pas lagi ngembangin aplikasi. Secara umum ada lima tahapan dalam pengembangan aplikasi yaitu:
1. Requirement (Analisa Kebutuhan), di tahap ini tim member bakal melakukan riset ke user biar tau aplikasi dan sistem seperti apa yang mau dibuat.
2. Design, tahap dimana tim design akan membuat interface (tampilan) aplikasi.
3. Development, dimana tim member akan mengimplementasikan code-code/script pemograman supaya desain yang dibuat bisa menjadi aplikasi yang interaktif.
4. Testing, tahap ini dilakukan supaya bisa mengetahui apakah aplikasi sudah bisa dijalankan dengan baik sesuai harapan user atau belum. Pengujian bisa diakukan oleh tim, bisa juga dengan melibatkan user.
5. Deployment, Tahap dimana aplikasi udah release dan dapat dipakai oleh user.

SDLC memiliki berbagai pendekatan, dua diantaranya yaitu:
1. Waterfall, dinamakan waterfall karena setiap tahapnya kayak air terjun, dimana setiap tahap bakal dijalanin secara linear atau lurus. Mulai dari tahap awal sampai akhir. Pendekatan ini punya kekurangan kurang fleksibel. Apabila ada feedback dari user, kita nggak bisa mengulang lagi prosesnya dari awal, karena prosesnya linear.
2. Agile, punya tahapan yang iteratif alias berulang. Yang berarti, pas selesai pengembangan, bakal selalu ada evaluasi, perbaikan, riset, plan buat fitur baru. Dengan begitu pula, apabila ada feedback, aplikasi yang dibuat dapat menyesuaikan. Agile punya sebuah framework yang paling sering dipakai yaitu Scrum.

### 4.2 Framework Scrum

<img src="https://i.ibb.co/Yh9StW6/Whats-App-Image-2022-07-20-at-6-25-40-AM.jpg" width="500">  

Scrum bakal dibagi menjadi bebrapa batasan waktu (timebox) yang disebut sebagai sprint. Biasanya satu sprint berduarasi 1 sampai 4 minggu. Hal itu dilakukan supaya proses negmbangin aplikasi bisa lebih cepat nerima feedback dari user dan lebih efektif buat diselesaikan. Berikut tata cara penerapan scrum:
1. Setiap mau mulai sprint, perlu ada perencanaan dulu. Tujuannya biar sprint bisa berjalan dengan baik. Nah tahap perencanaan itu disebut sprint planning. Dimana tim bakal mendefinisikan atau nentuin semua sprint goalnya.
- Sprint goal yang udah dibuat, selanjutnya bakal dipecah jadi task-task kecil. Sprint backlog adalah keluaran sprint planning, yaitu daftar pekerjaan hasil kesepakatan Product Ownber dan Dvelopment Team selama satu Sprint. 

<img src="https://i.ibb.co/qjsVpxf/Whats-App-Image-2022-07-20-at-6-25-40-AM-1.jpg" width="500">

- Tim juga perlu pertimbangin kemampuan diri pas ngerjain task. Misalnya, tim memutuskan buat mecah jadi 4 task; task A, B, C, D. Pas sprint pertama, tim sepakat cuma bisa ngerjain 3 task terpenting yaitu task A, B, dan C. Setelahnya,. ketiga task itu bakal dipindahin dari product backlog ke sprint backlog.
- Pas sebuah task mulai dikerjaiin, task ity masuk ke dalam kolom in progress. Pas selesai, taskdari in progress masuk ke kolom done. Nah, task yang belum sempet dikerjain tetap ada di kolom sprint backlog, yang nantinya bakal dikerjain di sprint selanjutnya. 
2. Setelah sprint planning di hari pertama, selanjutnya setiap hari bakal ada daily scrum atau dikenal dengan istilah daily stand up. Tujuannya agar tim slaing tau progress satu sama lain dan saling bantu apabila ada kendala. Disini juga dilihat apakah tim masih on track atau tidak. Durasinya selama 15 menit. Bedanya dengan sprint planning, di daily scrum kita nggak bisa ubah task karena sprint udah jalan. Kita cuma bisa lapora terkait kerjaan yang sudah selesai, kendala, dan mau ngapain hari ini.
3. Setiap kelar sprint, akan ada sprint review. Stake holder akan diundang, tujuannya buat dapein feedback dari berbagai perspekrtif demi perkembagan produk supaya makin baik.
4. Setelah hasil sprint dibahas di sprint review, selanjutnya ada sesi sprint retrospective yang mana tim bakal merefleksikan proses scrum yang udah dijalanin. Mendiskusikan bersama tim apa yang perlu diperbaiki, dipertahanin, dan ditingkatkan. Hasil diskusi tersebut, buat ningkatin efektifitas proses scrum di sprint selanjutnya.

### 4.3 Scrum Member

1. Team Member atau Developer, terdiri dari UI/UX Designer, Software Engineer, QA Engineer, dll.
2. Product Owner, orang yang berperan nentuin prioritas tugas-tugas yang harus diselesaikan duluan oleh team member. Setiap tim berhak nentuin jumlah tugas sesuai kemampuan mereka. 
3. Scrum Master, orang yang paham banget gimana tata cara berjalannya proses scrum.  Memiliki tugas utama buat memantau kelancaran pengembangan produk biar tetap sesuai tata caranya Scrum. Dan harus memilki kemampuan mentorship supaya bisa membantu menyelesaikan masalah internal antar team member.

## 5. Memahami Team Member Scrum

### 5.1 Berbagi Peran dalam Proses
1. UI/UX Designer
- UX (User Experience), bertanggung jawab buat memastikan pengalaman dan kenyamanan user pas pakai aplikasi. Idealnya melakukan riset terlebih dahulu untuk mendapatkan flow (alur) yang menyenangkan user yang dapat dilakukan dalam bentuk wawancara, survey, pengamatan perilaku sehari-hari.
- UI (user Interface), design sendiri ada 2 tahapan yaitu lo-fi design dan hi-fi design. Lo-fi (low fidelity design) adalah design yang udah bisa gambarin flow user dan letak-letak button dan text, cuman belum secandtik design beneran. Di tahap inilah, sebagai pengujian design ke user pula, apabila ingin direvisi-revisi. Adapun hi-gi (high fidelity design) adalah output final design yang dibuat dimana harus intuituf, artinya user harus gampang pake fitur-fitur tanpa dijelasin.

<img src="https://i.ibb.co/cxwr6Bq/Whats-App-Image-2022-07-20-at-9-51-32-AM.jpg" width="500">

2. Software Engineer, terbagi menjadi dua:
- Front Engineer
- Backend Engineer, secara umum ada 3 bagian penting yaitu; server, database, dan API. Didalam server, tersimpan kumpulan data yang disebut database. Server dan database perlu di manage agar data tersimpan dengan aman. Adapun API bertugas sebagai pengirim dan pemberi data. 
Output yang dibikin dari hasil kolaborasi UI/UX Designer, Backend Engineer, dan Frontend Engineer disebut staging app. Staging app sederhananya adalah interface yang interaktif dan bisa input maupun request data ke server.
3. QA Engineer, tugasnya memastikan aplikasi yang dirilis minim masalah, seperti bug dan error. Testing yang dilakukan dibagi menjadi dua manual dan automated. Manual testing cocok dilakukan buat test yang bener-bener baru dan test apikasi yang masih sering ada perubahan. Sedangkan automated testing cocok buat test yang sekiranya bakal melelahkan kalo dilakuin secara manual, contohnya testing situs multilingual (pake banyak bahasa). Setiap tes harus didokumentasikan. Banyak dari QA Enginerr yang akhirnya jadi seorang product owner karena dia tahu luar-dalem permasalah yang terjadi di SDLC dan tau cara menimalisirnya, ngerti bahasa pemograman yang dipakai di pembuatan API oleh backend engineer, perhatian terhadap detail dan tahu cara dokumentasi teknis yang baik.
